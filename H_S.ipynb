```python
import os
from typing import TypedDict

from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
```

```python
# ============================================================
# LLM
# ============================================================
```

```python

llm = ChatOpenAI(
    model="gpt-5.4",                     # your custom model name
    temperature=0.3,
    openai_api_key="sk-............",
    openai_api_base="https://api.gapgpt.app/v1"
)
```

```python
# ============================================================
# STATE
# ============================================================
```

```python
class StudyState(TypedDict):
    course_material: str
    exam_days: int
    daily_hours: int

    course_analysis: str
    concepts: str
    dependencies: str
    assessment: str
    time_allocation: str
    strategies: str
    study_plan: str
    quizzes: str
    final_report: str
```

```python
# ============================================================
# NODE 1: COURSE ANALYSIS
# ============================================================
```

```python
def analyze_course(state: StudyState) -> StudyState:

    prompt = ChatPromptTemplate.from_template("""
    You are a university professor.

    Analyze this course material.

    Extract:
    - Main topics
    - Difficulty levels
    - Important concepts
    - Exam-critical sections

    COURSE:
    {course_material}
    """)

    result = (prompt | llm).invoke({
        "course_material": state["course_material"]
    })

    state["course_analysis"] = result.content
    return state
```

```python
# ============================================================
# NODE 2: CONCEPT EXTRACTION
# ============================================================
```

```python
def extract_concepts(state: StudyState) -> StudyState:

    prompt = ChatPromptTemplate.from_template("""
    Extract all major concepts from:

    {analysis}

    For each concept include:
    - Name
    - Importance
    - Short explanation
    """)

    result = (prompt | llm).invoke({
        "analysis": state["course_analysis"]
    })

    state["concepts"] = result.content
    return state
```

```python
# ============================================================
# NODE 3: DEPENDENCY MAPPING
# ============================================================
```

```python
def map_dependencies(state: StudyState) -> StudyState:

    prompt = ChatPromptTemplate.from_template("""
    Build a learning dependency map.

    Concepts:
    {concepts}

    Determine:
    - Learning order
    - Prerequisites
    """)

    result = (prompt | llm).invoke({
        "concepts": state["concepts"]
    })

    state["dependencies"] = result.content
    return state
```

```python
# ============================================================
# NODE 4: STUDENT ASSESSMENT
# ============================================================
```

```python
def assess_student(state: StudyState) -> StudyState:

    prompt = ChatPromptTemplate.from_template("""
    Assume a typical student.

    Concepts:
    {concepts}

    Identify:
    - Strengths
    - Weaknesses
    - High-risk topics
    """)

    result = (prompt | llm).invoke({
        "concepts": state["concepts"]
    })

    state["assessment"] = result.content
    return state
```

```python
# ============================================================
# NODE 5: TIME ALLOCATION
# ============================================================
```

```python
def allocate_time(state: StudyState) -> StudyState:

    total_hours = state["exam_days"] * state["daily_hours"]

    prompt = ChatPromptTemplate.from_template("""
    Create a study-hour allocation.

    Total Hours: {hours}

    Concepts:
    {concepts}

    Assessment:
    {assessment}
    """)

    result = (prompt | llm).invoke({
        "hours": total_hours,
        "concepts": state["concepts"],
        "assessment": state["assessment"]
    })

    state["time_allocation"] = result.content
    return state
```

```python
# ============================================================
# NODE 6: STRATEGIES
# ============================================================
```

```python
def select_strategies(state: StudyState) -> StudyState:

    prompt = ChatPromptTemplate.from_template("""
    Choose best learning strategies:

    - Flashcards
    - Active Recall
    - Practice Problems
    - Spaced Repetition
    - Mock Exams

    Concepts:
    {concepts}
    """)

    result = (prompt | llm).invoke({
        "concepts": state["concepts"]
    })

    state["strategies"] = result.content
    return state
```

```python
# ============================================================
# NODE 7: STUDY PLAN
# ============================================================
```

```python
def generate_study_plan(state: StudyState) -> StudyState:

    prompt = ChatPromptTemplate.from_template("""
    Create a detailed study schedule.

    Days until exam: {days}

    Time Allocation:
    {allocation}

    Dependencies:
    {dependencies}

    Include revision sessions.
    """)

    result = (prompt | llm).invoke({
        "days": state["exam_days"],
        "allocation": state["time_allocation"],
        "dependencies": state["dependencies"]
    })

    state["study_plan"] = result.content
    return state
```

```python
# ============================================================
# NODE 8: QUIZ GENERATION
# ============================================================
```

```python
def generate_quizzes(state: StudyState) -> StudyState:

    prompt = ChatPromptTemplate.from_template("""
    Create 10 exam-style questions.

    Concepts:
    {concepts}

    Include:
    - MCQs
    - Short answers
    - Difficulty levels
    """)

    result = (prompt | llm).invoke({
        "concepts": state["concepts"]
    })

    state["quizzes"] = result.content
    return state
```

```python
# ============================================================
# NODE 9: FINAL REPORT
# ============================================================
```

```python
def create_final_report(state: StudyState) -> StudyState:

    prompt = ChatPromptTemplate.from_template("""
    Create a final personalized learning plan.

    ANALYSIS:
    {analysis}

    CONCEPTS:
    {concepts}

    DEPENDENCIES:
    {dependencies}

    ASSESSMENT:
    {assessment}

    TIME ALLOCATION:
    {allocation}

    STRATEGIES:
    {strategies}

    STUDY PLAN:
    {study_plan}

    QUIZZES:
    {quizzes}
    """)

    result = (prompt | llm).invoke({
        "analysis": state["course_analysis"],
        "concepts": state["concepts"],
        "dependencies": state["dependencies"],
        "assessment": state["assessment"],
        "allocation": state["time_allocation"],
        "strategies": state["strategies"],
        "study_plan": state["study_plan"],
        "quizzes": state["quizzes"]
    })

    state["final_report"] = result.content
    return state
```

```python
# ============================================================
# PIPELINE RUNNER (NO GRAPH)
# ============================================================
```

```python
def run_pipeline(state: StudyState) -> StudyState:

    state = analyze_course(state)
    state = extract_concepts(state)
    state = map_dependencies(state)
    state = assess_student(state)
    state = allocate_time(state)
    state = select_strategies(state)
    state = generate_study_plan(state)
    state = generate_quizzes(state)
    state = create_final_report(state)

    return state
```

```python
# ============================================================
# MAIN
# ============================================================
```

```python
if __name__ == "__main__":

    material = input("\nPaste course material:\n\n")

    state: StudyState = {
        "course_material": material,
        "exam_days": 30,
        "daily_hours": 2,

        "course_analysis": "",
        "concepts": "",
        "dependencies": "",
        "assessment": "",
        "time_allocation": "",
        "strategies": "",
        "study_plan": "",
        "quizzes": "",
        "final_report": ""
    }

    result = run_pipeline(state)

    print("\n" + "=" * 80)
    print("PERSONALIZED LEARNING PLAN")
    print("=" * 80)
    print(result["final_report"])
```
